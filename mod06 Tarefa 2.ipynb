####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carregar o arquivo CSV
file_path = './sinasc_RO_2019.csv'
df = pd.read_csv(file_path)

# Criar um boxplot da idade da mãe
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['IDADEMAE'])
plt.title('Boxplot da Idade das Mães')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Criar um boxplot do peso do bebê
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['PESO'])
plt.title('Boxplot do Peso dos Bebês')
plt.xlabel('Peso do Bebê (gramas)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Criar um histograma da idade da mãe
plt.figure(figsize=(10, 6))
plt.hist(df['IDADEMAE'].dropna(), bins=20, edgecolor='black')
plt.title('Histograma da Idade das Mães')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Criar um histograma do peso do bebê
plt.figure(figsize=(10, 6))
plt.hist(df['PESO'].dropna(), bins=20, edgecolor='black')
plt.title('Histograma do Peso dos Bebês')
plt.xlabel('Peso do Bebê (gramas)')
plt.ylabel('Frequência')
plt.show()
```
